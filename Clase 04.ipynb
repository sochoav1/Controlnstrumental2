```python
from sympy import *
init_printing()
```

```python
vi,iv,rv,vb,iu,ru,vd,r2,r3 = symbols("V_i,i_v,R_v,V_b,i_u,R_u,V_d,R_2,R_3")
```

```python
eq1 = Eq(vi-iv*rv-vb,0)
eq2 = Eq(vi-iu*ru-vd,0)
eq3 = Eq(iv*(rv+r2),vi)
eq4 = Eq(iu*(ru+r3),vi)
```

```python
display(eq1)
display(eq2)
display(eq3)
display(eq4)
```

```python
sol = solve([eq1,eq2,eq3,eq4],(iu,iv,vd,vb))
sol
```

```python
vo = symbols("V_o")
```

```python
eq5 = Eq(vo,vb-vd).subs(sol)
display(eq5)
```

```python
Eq(rv,solve(eq5.subs({vo:0}),rv)[0])
```
